```python


with open("input.txt", "r") as f:
    reports = [
        list(map(int, line.split()))
        for line in f
        if line.strip()
    ]

print("Number of reports:", len(reports))
print("First 5 reports:")
for r in reports[:5]:
    print(r)
```

```python
def is_safe(report):
    increasing = True
    decreasing = True
    step_ok = True

    for i in range(len(report) - 1):
        diff = report[i + 1] - report[i]

        if diff <= 0:
            increasing = False

        if diff >= 0:
            decreasing = False

        if abs(diff) < 1 or abs(diff) > 3:
            step_ok = False

    return (increasing or decreasing) and step_ok


safe_count = 0

for report in reports:
    if is_safe(report):
        safe_count += 1

print("Number of safe reports:", safe_count)
```

```python
def is_safe2(report):
    if is_safe(report):
        return True

    for i in range(len(report)):
        temp = report[:i] + report[i + 1:]

        if is_safe(temp):
            return True

    return False


safe_count2 = 0

for report in reports:
    if is_safe2(report):
        safe_count2 += 1

print("Number of safe reports with Problem Dampener:", safe_count2)
```
